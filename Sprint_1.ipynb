### Creating a DataFrame

```python
df = pd.concat([data_0,data_1,data_2,data_3,data_4,data_5,data_6,data_7,data_8,data_9])
df
```

### Machine Learning

```python
# inputs and target from dataframe
inputs = df.select_dtypes(include='float64')
target = df['labels']
```

### Splitting the data into train and test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(inputs, target, train_size=0.75, random_state =100)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression()

classifier.fit(X_train, y_train)
```

#### Prediction

```python
y_test_pred = classifier.predict(X_test)
```

```python
temp_df = pd.DataFrame({'Actual': y_test, 'Predicted':  y_test_pred})
temp_df.head()
```

#### Evaluation

```python
from sklearn.metrics import accuracy_score
score_1=accuracy_score(y_test, y_test_pred)
score_1
```
