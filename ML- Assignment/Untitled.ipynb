$ Machine\ Learning $

```python
# identifying input and output/target
target = df['Species']
inputs = df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
```

```python
# Identifying the ML Task
# Here the target variables is given so, it is a supervised learning.
# Here the target variable is a categorical or discrete column so, it is a Classification Task.
df.head()
```

$ Separating\ Categorical\ and\ Numerical\ Column $

```python
df_num = inputs.select_dtypes(include='float64')
df_num.head()
```

$ Applying\ LabelEncoding\ on\ Categorical\ Column $

```python
label_encoder = {'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2}

target = df['Species'].apply(lambda x: label_encoder[x])

target.head()
```

```python
# standardization of numerical column
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df_num = pd.DataFrame(scaler.fit_transform(df_num), columns = df_num.columns)

df_num.head()
```

$ Splitting\ data\ in\ to\ Train\ and\ Test\ DataFrame $

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(inputs, target, train_size=0.75, random_state=100)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

$ Logistic\ Regression $

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression()

classifier.fit(X_train, y_train)
```

```python
y_test_pred = classifier.predict(X_test)
```

```python
temp_df = pd.DataFrame({'Actual': y_test, 'Predicted':  y_test_pred})
temp_df.head()
```

```python
#from sklearn import metrics
#print('Mean Absolute Error: ', metrics.mean_absolute_error(y_test, y_test_pred))

#print('Mean squared Error:', metrics.mean_squared_error(y_test, y_test_pred))

#print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_test_pred)))
```

```python
from sklearn.metrics import accuracy_score
score_1=accuracy_score(y_test, y_test_pred)
score_1
```

```python
y_test_pred = classifier.predict(X_test)
```

```python
temp_df = pd.DataFrame({'Actual': y_test, 'predicted': y_test_pred})

temp_df.head()
```
